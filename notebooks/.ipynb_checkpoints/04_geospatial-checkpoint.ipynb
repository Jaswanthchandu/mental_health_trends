```python
map_df = df[df['YearStart'] == 2022].groupby(['LocationAbbr', 'LocationDesc'])['DataValue'].mean().reset_index()

fig = px.choropleth(map_df,
                    locations='LocationAbbr',
                    locationmode='USA-states',
                    color='DataValue',
                    scope='usa',
                    color_continuous_scale='Reds',
                    title='Mental Health Prevalence by State (2022)',
                    labels={'DataValue': 'Prevalence Rate (%)', 'LocationAbbr': 'State'},
                    hover_name='LocationDesc',
                    hover_data={'DataValue': ':.1f', 'LocationAbbr': False})

fig.add_scattergeo(
    locations=map_df['LocationAbbr'],
    locationmode='USA-states',
    text=map_df['LocationAbbr'],
    mode='text',
    textfont=dict(size=10, color='black'),
    showlegend=False
)

# Make map bigger and cleaner
fig.update_layout(
    geo=dict(scope='usa'),
    width=1100,
    height=600,
    title_font_size=18,
    margin=dict(l=0, r=0, t=50, b=0)
)

fig.show()
```

```python
fig = px.choropleth(df[df['YearStart'].isin([2019, 2022])].groupby(['YearStart', 'LocationAbbr'])['DataValue'].mean().reset_index(),
                    locations='LocationAbbr',
                    locationmode='USA-states',
                    color='DataValue',
                    scope='usa',
                    animation_frame='YearStart',
                    color_continuous_scale='Reds',
                    title='Mental Health Prevalence by State (2019 vs 2022)',
                    labels={'DataValue': 'Prevalence Rate (%)'})

fig.update_layout(width=1100, height=600)
fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# Get 2019 and 2022 data
df_2019 = df[df['YearStart'] == 2019].groupby('LocationAbbr')['DataValue'].mean().reset_index()
df_2022 = df[df['YearStart'] == 2022].groupby('LocationAbbr')['DataValue'].mean().reset_index()

# Create side by side subplots
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Pre-COVID (2019)', 'During COVID (2022)'),
    specs=[[{'type': 'choropleth'}, {'type': 'choropleth'}]]
)

# 2019 map
fig.add_trace(go.Choropleth(
    locations=df_2019['LocationAbbr'],
    z=df_2019['DataValue'],
    locationmode='USA-states',
    colorscale='Reds',
    zmin=8, zmax=20,
    showscale=False,
    name='2019'
), row=1, col=1)

# 2022 map
fig.add_trace(go.Choropleth(
    locations=df_2022['LocationAbbr'],
    z=df_2022['DataValue'],
    locationmode='USA-states',
    colorscale='Reds',
    zmin=8, zmax=20,
    colorbar_title='Prevalence Rate (%)',
    name='2022'
), row=1, col=2)

fig.update_geos(scope='usa')
fig.update_layout(
    title_text='Mental Health Prevalence: Pre-COVID (2019) vs During COVID (2022)',
    title_font_size=16,
    width=1200,
    height=500,
    margin=dict(l=0, r=0, t=60, b=0)
)

fig.show()

# Save both versions
fig.write_html('../visuals/comparison_map.html')
fig.write_image('../visuals/comparison_map.png')
print("Saved!")
```
